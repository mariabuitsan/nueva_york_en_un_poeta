```python
import nltk 
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
#accesing my local .txt file 

textFGL = open ('Poeta_en_Nueva_York-cleaned3.txt', 'r')
```

```python
#saving textFGL as a new variable and reading

projectFGL = textFGL.read()
print(projectFGL)
```

```python
type(projectFGL)
```

```python
#Using a function to split strings with "," as the delimeter

def convert(string):
    li = list(string.split(","))
    return li
```

```python
#calling the function

listFGL = convert(projectFGL)
```

```python
print(listFGL)
```

```python
#do I finally have a list!!? 

type(listFGL)
```

```python
#making my list all lowercase

lowercaseFGL = []

for item in listFGL:
    lowercaseFGL.append(item.lower())
```

```python
lowercaseFGL
```

```python
#saving lowercaseFGL as an nltk object

wordsFGL = nltk.Text(lowercaseFGL)
```

```python
#testing 

wordsFGL.count('caballos')
```

```python
#testing again...I need to clean my text better because I have a different word count with VSC

wordsFGL.count('caballo')
```

```python
#deleting stopwords, tokenizing and lemmatizing using code from class

cleanFGL = []
stops = stopwords.words('spanish')

for token in wordsFGL:
    if token.isalpha():
        if token not in stops:
            lemmatizer = WordNetLemmatizer()
            lemma = lemmatizer.lemmatize(token)
            cleanFGL.append(lemma.lower())
```

```python
#glimpse 1 looks good
cleanFGL[:20]
```

```python
#glimpse 2 ...hmm not so good, 'ruiseñorlloraba' is not a word! ):

cleanFGL[4521:4560]
```

### Conclusion



I was surprised to find that 'love' is the most used word in the text given that the majority of the poems are denouncing the american industrial capitalist society of the 1930s. It might be because FGL in this text is constantly naming his love, as a noun, as an object, rather than using 'love' as a verb. For example, the last time he uses the word 'horse' in the collection is also to describe what 'his love' is not ('and my love that is not a horse'). So, even though 'Poet in New York' is a critique to modern industrial society it is also, as many authors have shown, an exploration of his own persona, his inner world reacting to the city. Apparently he once said that the tittle of the book should've been "New York in a Poet" rather than "Poet in New York". He's also constantly reflecting about children and childhood with poems like "Tu infancia en mentón" or "El niño Stanton" which is why "niño" is among the list of most used words. It is also interesting that the collocation method returns "childhood and fable" as co-occurring, which might just come from the first line of a poem containing both words, but then again it can also suggest that childhood belongs to the same semantic universe of things that for FGL are mythic. The other most used words are definetly to be expected: 'blood', 'dead','pain'. Now, his use of animals, with 'horse' as both one of the most used words and the most used animal is interesting. My partial hypothesis after this exercise is that FGL uses each animal with a very different and particular symbolic intention. Allen Josephs explains that FGL uses 'dove' as a religious symbol to denounce american materialism, which explains why 'christ' appears next to it and also 'skulls' are in the same context as dove. Dove might have the poetic function of visualizing the dying religious morals of the american society, which supports Joesphs argument. But it's only a poetic method used by a secular voice, FGL's complaint is more about the loss of any kind of mythic vision than a religious protest. In contrast, I think horses belong to FGL's inner mythic world and that's why they appear in relation to words like "to be" (ser) or "craziness" (locura) or "effusive" (efusivo) or "to dream" (soñar), etc. His horses tend to be blue and white. His horses tend to be named along with possesives, which are deleted as stopwords. His horses also appear next the first person pronoun, which I think are also deleted. Maybe horses are personal and important because they are also part of the Andalusian landscape, where he's from. I don't know. Although horses are certaintly symbolic of the Spanish imaginary, like the horse in Piccaso's Guernica. I would like to visualize the 15 horses that FGL names in Poet in New York and store them in a list. I would like to do this exercise for other less known authors. But for a better analysis I need to clean the text more. I was disappointed whit the limited words surrounding "whitman" which is one of the most important poems in the collection. I was expecting 'river' and 'butterflies' and got nothing. Finally, collocations returns 'verdadero'(true/real) and 'dolor' (pain) as co-occurring words, I think this could support an argument about FGL's suffering while writing the poems or while living in NYC. And is so sad to think he went back to Spain and was assesinated three years after writing this in his own natal town.
